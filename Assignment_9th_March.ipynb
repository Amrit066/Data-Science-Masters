##### Part-1:<br><br>
The normal distribution is frequently used to comprehend population factor distributions. The normal distribution has evolved into a standard of reference for many probability issues because it accurately approximates a wide range of natural phenomena..<br><br>
For example,<br><br>
- People's heights serve as an illustration of regular distribution. A population's majority of members are of normal height. A very small percentage of people are either extremely tall or extremely short, and the number of people who are both taller and shorter than normal people is nearly equal. Height is influenced by a number of genetic and environmental variables. It thus adheres to the normal distribution.
- Another excellent illustration of a normal distribution is a fair toss of the dice. In an experiment, it was discovered that the odds of getting "1" are 15–18% when rolling a die 100 times, and they remain the same when rolling it 1000 times, averaging 16.7% (1/6). There are 36 different outcomes if we throw two dice at once. Again, the average chance of rolling a "1" (with six potential combinations) is about 16.7%, or 6/36. More the number of dice more elaborate will be the normal distribution graph.<br><br>
##### Part-2:<br><br>
The two main parameters of a (normal) distribution are the mean and standard deviation. The parameters determine the shape and probabilities of the distribution. The shape of the distribution changes as the parameter values change.<br><br>
- In a normal distribution graph, the mean defines the location of the peak, and most of the data points are clustered around the mean. Any changes made to the value of the mean move the curve either to the left or right along the X-axis.
- The standard deviation measures the dispersion of the data points relative to the mean. It determines how far away from the mean the data points are positioned and represents the distance between the mean and the observations. - The central limit theorem states that the distribution of a sample variable approximates a normal distribution as the sample size becomes larger, assuming that all samples are identical in size, and regardless of the population's actual distribution shape.
<br><br>
##### Part-2: Significance of CLT are;<br><br>
- The central limit theorem is helpful when analyzing large data sets.
- It enables one to assume that the sampling distribution of the mean will typically be normally distributed. This makes statistical research and inference simpler.
