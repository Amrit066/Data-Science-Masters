- We usse the url for() function to create a URL to a certain function dynamically. 
- It takes any number of keyword parameters, each of which corresponds to a different variable part of the URL rule, as well as the function's name as its first argument. 
- The URL's query parameters are inserted as unknown variable sections.
<br><br>
##### Part-2: Example;<br>

- Importing required libraries:

```python
# Importing Libraries
from flask import Flask
from flask import url_for
from flask import redirect
```

- Creating Routes:

```python
# Specific Route With Input

@app.route("/home")
def home():
    return "<h1>This is home page.</h1>"

@app.route("/welcome")
def welcome_msg():
    return "<h1>Welcome to ABC Corporation</h1>"
```

- Using __url_for()__ to create dynamic routes and then calling those routes:

```python
@app.route("/<utype>")
def url_type(utype):

    if(utype == "home"):
        return redirect(url_for("home")) # Here, a URL for "/home" is created
                                         # and then redirected to it.
    
    if(utype == "welcome"):
        return redirect(url_for("welcome")) # Here, a URL for "/home" is created
                                         # and then redirected to it.
```
