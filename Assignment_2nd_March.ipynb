# 2.

##### Part-1:<br><br>
- A scatter plot is a set of points plotted on a horizontal and vertical axes. 
- It is used to show the relationship between different variables.
<br><br>
##### Part-2:<br>

```python
# Generating data for x and y co-ordinates
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
# Plotting Data
plt.figure(figsize=(5,4))
plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()
```

# 3.

##### Part-1:<br><br>
- subplot() function is used to draw multiple plots in a single figure.
- It adds subplot to a current figure at the specified grid position.
<br><br>
##### Part-2:<br><br>

```python
# Creating data for each subplot
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
# Plottng Data
plt.subplot(2,2,1)
plt.plot(x1,y1,c="#FB2576")
plt.xlabel("X1-axis")
plt.ylabel("Y1-axis")
plt.title("Line Plot-1")

plt.subplot(2,2,2)
plt.plot(x2,y2,c="#00FFD1")
plt.xlabel("X2-axis")
plt.ylabel("Y2-axis")
plt.title("Line Plot-2")

plt.subplot(2,2,3)
plt.plot(x3,y3,c="#5800FF")
plt.xlabel("X3-axis")
plt.ylabel("Y3-axis")
plt.title("Line Plot-3")

plt.subplot(2,2,4)
plt.plot(x4,y4,c="#247881")
plt.xlabel("X4-axis")
plt.ylabel("Y4-axis")
plt.title("Line Plot-4")

# Adjudting the size and spaces between subplots
plt.subplots_adjust(right=1.8, top=1.8)

plt.show()
```

# 4.

##### Part-1:<br><br>
- A bar graph is a graphical representation of information. 
- It uses bars that extend to different heights to depict value.
- It can have an x-axis and a y-axis and can be used to compare one or more categories of data.
<br><br>
##### Part-2:<br><br>
- It is used to  show the relationship between a numeric and a categoric variable. 
- Each entity of the categoric variable is represented as a bar. And the size of the bar represents its numeric value.
<br><br>
##### Part-3:<br>

```python
# Creating Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
# Plotting Data

# Vertical Bar Plot
plt.subplot(2,1,1)
plt.bar(company,profit,color="#247881")

# Horizontal Bar Plot
plt.subplot(2,1,2)
plt.barh(company,profit,color="#5800FF")
```

# 5.

##### Part-1:<br><br>
- Box plot is a chart that shows data from a five-number summary.
- It shows the minimum, maximum, first quartile, third quartile, median and outliers.
<br><br>
##### Part-2: It is used to know:<br><br>
- The outliers and their values
- Symmetry of Data
- Tight grouping of data
- Skewness of data
<br><br>
##### Part-3:<br>

```python
# Creating Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
# Plotting Data
plt.boxplot([box1,box2],patch_artist=True,labels =["Box-1","Box-2"])
plt.show()
```
