```python
# Importing Libraries
from flask import Flask
from flask import request
from flask import jsonify

# Creating Flask Object
app = Flask(__name__)

#  Creating route

@app.route("/using_postman", methods=["POST"])
def display_name():

    case = request.json["Case"]
    fname = request.json["First Name"] 
    lname = request.json["Last Name"]
    res= ""

    if(case.lower() == "show first name"):
        res = "First name of the person is {}.".format(fname)
    if(case.lower() == "show last name"):
        res = "Last name of the person is {}.".format(lname)
    if(case.lower() == "show full name"):
        res = "Full name of the person is {} {}.".format(fname,lname)
    
    return res
```
